A class is a blueprint for a object which includes Methods and Attributes.

```python
# Instance variables and methods

# CREATING CLASS
class dog:
    # CONSTRUCTOR
    def __init__(self,name,age):
            self.name = name
            self.age = age
            
# CREATING OBJECT
d1 = dog("tommy",1)
print(d1.name)
print(d1.age)
```

```python
d2 = dog("fluffy",4)
print(d2.name)
print(d2.age)
```
