A class is a blueprint for a object which includes Methods and Attributes.

```python
# Instance variables and methods

# CREATING CLASS
class dog:
    # CONSTRUCTOR
    def __init__(self,name,age):
            self.name = name
            self.age = age
            
# CREATING OBJECT
d1 = dog("tommy",1)
print(d1.name)
print(d1.age)
```

```python
d2 = dog("fluffy",4)
print(d2.name)
print(d2.age)
```

```python
# Designing a class with attributes and mwthods

class dog:
    # constructor for attributes
    def __init__(self,name,age,color):
        self.name = name
        self.age = age
        self.color = color
    
    def barks(self):
        print(f"{self.name} barks like bhauuu bhauuuuuu")
        
    def rang(self):
        if self.color == "black":
            print(f"{self.name} ek kala kutta hai")
        elif self.color == "white":
            print(f"{self.name} ek safeed kutta hai")


dogo1 = dog("kutra",2,"black")
dogo2 = dog("fluffy",4,"white")
dogo1.barks()
dogo2.barks()
dogo1.rang()
dogo2.rang()
```

```python
# Modelling a banking system credit debit

class BankAccount:
    def __init__(self,owner,balance=0):
        self.owner = owner
        self.balance = balance
    
    def credit(self,incoming):
        self.balance = self.balance + incoming
        print(f"The new balance is {self.balance}")
        
    def debit(self,out):
        if out > self.balance:
            print("Insufficient Funds In The Bank Account")
        else:
            self.balance = self.balance - out
            print(f"The new balance is {self.balance}")
            
    def getBalance(self):
        print(f"The Account Balance is {self.balance}")
        
acc1 = BankAccount("Atharv",0)
acc2 = BankAccount("Ayush",0)

acc1.credit(12000)
acc2.credit(2000)

acc1.debit(3000)
acc2.debit(1000)

acc1.getBalance()
```
