#ENCAPSULATION

Encapsulation is the concept of wrapping data (variables) and methods (functions) together as a single unit. It restricts direct access to some of the object's components, which is a means of preventing accidental interference and misuse of the data.

PUBLIC variables → Accessible from anywhere.
PROTECTED variables → Intended for class + derived classes (but not enforced).
PRIVATE variables → Intended for class only (name-mangling prevents accidental access).

```python
class Person:
    def __init__(self,name,age,gender,color):
        self.__name = name      # private variable
        self.__age = age        # private variable
        self._color = color     #Protected Variable
        self.gender = gender   # public variable
def getName(person):
    print(person.__name)

def getAge(person):
    print(person.__age)
    
def getGender(person):
    print(person.gender)
    
person1 = Person("Atharv",21,"Male","White")
getGender(person1)  # Accessible

class Student(Person):
    def __init__(self,name,age,gender,color):
        Person.__init__(self,name,age,gender,color)

stu1 = Student("pankaj",12,"male","brown")

print(stu1._color)
   
```

```python
getName(person1)    # Not Accessible as it is private
getAge(person1)  # Not Accessible as it is private
```

```python
# ENCAPULATION WITH GETTER AND SETTERS

class Person:
    def __init__(self,name,age,gender):
        self.__name = name
        self.__age = age
        self.gender = gender
    
    # Getter method for name
    def get_name(self):
        return self.__name
    
    # Getter for age
    def get_age(self):
        return self.__age
    
    # SETTER FOR NAME
    
    def set_name(self,name):
        self.__name = name
    
    # SETTER FOR AGE
    
    def set_age(self,age):
        if age>0:
            self.__age = age
        else:
            print("Age cannot be less than 0")
        
p1 = Person("Atharv",21,"Male")

print(p1.get_age())
print(p1.get_name())

p1.set_name("Ayush")
print(p1.get_name())

p1.set_age(2)
print(p1.get_age())
```
