POLYMORPHISM 

It is a core concept in Oops that allows a object of different classes to be treated as a object of a common superclass.Polymorphism is achived through method overloading and method overriding

METHOD OVERRIDING
It allows a subclass or child class to provide a specific implementation of a method that is already provided by one of its super-classes or parent classes.

```python
class Animal:
    def speak(self):
        print("Animal speaks")
        
class Dog(Animal):
    def speak(self):
        return "bhau bhau"

class Cat(Animal):
    def speak(self):
        return "meow meow"
    
dog1 = Dog()
cat1 = Cat()
print(dog1.speak())
print(cat1.speak())
```

POLYMORPHISM USING ABSTRACT CLASSES


```python
from abc import ABC,abstractmethod

# DEFINING ABSTRACT CLASS
class Vehicle(ABC):
    @abstractmethod
    def engineStart(self):
        pass

# DERIVED CLASSES

class Car(Vehicle):
        def engineStart(self):
            print("Car engine started")
            
class Bike(Vehicle):
        def engineStart(self):
            print("Bike engine started")
            
# creating obj
car1 = Car()
Bike1 = Bike()

def printEngine(Vehicle):
    Vehicle.engineStart()
    
printEngine(car1)
printEngine(Bike1)
```
