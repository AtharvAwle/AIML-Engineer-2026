Inheritence in PY

It is a fundamental concept in Oops which allows us to inherit or use the attributes and methods from another class.
There are two types of inheritance SINGLE LEVEL  &  MULTIPLE LEVEL INHERITANCE


```python
# Parent Class

class Car:
    def __init__(self,windows,doors,engineType):
        self.windows = windows
        self.doors = doors
        self.engineType = engineType
        
    def drive(self):
        print(f"The person will drive a {self.engineType} car with {self.windows} windows and {self.doors} doors.")
```

```python
car1 = Car(4,5,"petrol")
car1.drive()
```

```python
# Child Class

class Tesla(Car):
    def __init__(self,windows,doors,engineType,isAutopilot):
        super().__init__(windows,doors,engineType)
        self.isAutopilot = isAutopilot
        
    def autopilotFeature(self):
        if self.isAutopilot:
            print("This car is self driving")
        else:
            print("This car is not self driving")
```

```python
tesla1 = Tesla(4,5,"electric",True)
tesla1.drive()
tesla1.autopilotFeature()
```

MULTIPLE INHERITANCE

```python
class Animals:
    def __init__(self,name):
        self.name = name
    def speak(self):
        pass
    
class Pet:
    def __init__(self,owner):
        self.owner = owner
        
class Dog(Animals,Pet):
    def __init__(self,name,owner,breed):
        Animals.__init__(self, name)
        Pet.__init__(self, owner)
        self.breed = breed
    def speak(self):
        return f"{self.name} says bhau bhau"
```

```python
dog1 = Dog("Banku","Atharv","Street")
dog2 = Dog("Tommy","John","Bulldog")
print(dog1.speak())
print(dog2.speak())
```
