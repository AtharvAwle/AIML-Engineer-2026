ITERATORS

Iterators are advance python concept that are used for efficient looping and memory management

iter() --> to initialise the iterator
next() --> to access the element in the list

```python
l1 = [1,2,3,4,5,6]
```

```python

iterator = iter(l1) #initializing the iterator
next(iterator)  #accessing the elements one by one
```

```python
# string iterator
s = "Atharv"
string_iterator = iter(s)
try:
    next(string_iterator)
except StopIteration:
    print("End of String reached")
      #StopIteration Error
```
