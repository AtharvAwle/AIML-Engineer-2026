File Operation- Read And Write Files
File handling is a crucial part of any programming language. Python provides built-in functions and methods to read from and write to files, both text and binary.

```python
# Read The Whole File

with open('example.txt','r') as file:
    content = file.read()
    print(content)
```

```python
# Read A File Line By Line
with open('example.txt','r') as file:
    for line in file:
        print(line.strip())  # Using strip() to remove leading/trailing whitespace
```

```python
# write a file(overwrite the existing content)
with open('example.txt','w') as writeFile:
    writeFile.write("This is a new line added to the file.\n")
    writeFile.write("This file is created using Python.\n")
    
```

```python
with open("example.txt","a") as appendFile:
    appendFile.write("Hey this is atharv awle I am learning AI ML And also Data Science \n")
```

```python
# writing the list of lines
lines = ["first \n" , "second \n","third"]
with open("example.txt","a") as file:
    file.writelines(lines)
```

READ AND WRITE A BINARY FILE IN PYTHON

```python
data = b'\x00\x01\x02\x03\x04'
with open('example.bin','wb') as file:
    file.write(data)
```

```python
with open('example.bin','rb') as file:
    content = file.read()
    print(content)
```

```python
# Read a file from one file and write it in another destination file

with open('example.txt','r') as source:
    content = source.read()
    
with open('destination.txt','w') as destination:
    destination.write(content)
```
